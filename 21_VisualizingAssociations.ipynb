```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df2 = pd.DataFrame([[7.1, 3.4, 0], [6.7, 4.0, 0], [5.6, 9.3, 1],
                    [5.3, 7.8, 1], [5.8, 4.6, 2], [2.5, 3.2, 2]],
                   columns = ['length', 'width', 'types'])

display(df2)
df2.plot.scatter(x = 'length', y='width', c = 'Blue')
df2.plot.scatter(x = 'length', y='width', c='types')
df2.plot.scatter(x = 'length', y='width', marker='^', s = 60, colormap = 'plasma')
df2.plot.scatter(x = 'length', y='width', s = df2['types'] * 50)
```

```python
import random

df3= pd.DataFrame(np.random.rand(20, 4) * 5, columns = ['a', 'b', 'c', 'd'])
display(df3)

ax=df3.plot.scatter(x= 'a', y='b', color='Skyblue', label='Group 1')
ax = df3.plot.scatter(x='c', y='d', color = 'Orange', label='Group 2', ax=ax)
plt.show()
```

```python
df4 = pd.DataFrame(np.random.rand(20, 4) * 5, columns = ['a','b','c','d'])

df4['class'] = np.round(np.random.rand(20, 1) *5)
df4['class'] = df4['class'].astype('category')

display(df4)

pd.plotting.scatter_matrix(df4, s= 60, diagonal='kde', c = df4['class'], cmap = 'spring')
pd.plotting.scatter_matrix(df4, s= 60, diagonal='hist', c = df4['class'], cmap = 'viridis')
```

```python
import seaborn as sns

iris = sns.load_dataset('iris')
```

```python
iris.plot.scatter(x = 'sepal_length', y = 'sepal_width')
plt.title('Scatter Plot(iris)', fontsize = 15)
plt.xlabel('Sepal Length', fontsize = 12)
plt.ylabel('Sepal Width', fontsize = 12)
plt.show()
display(iris)
```

```python
g = iris.groupby('species')

for name, group in g:
    plt.scatter(group.sepal_length, group.sepal_width, label = name)
plt.legend(fontsize = 12, loc = 'upper left')
plt.title('Scatter Plot(iris)', fontsize = 15)
plt.xlabel('Sepal Length', fontsize = 12)
plt.ylabel('Sepal Width', fontsize = 12)
plt.show()


for name, group in g:
    plt.scatter(group.petal_length, group.petal_width, label = name)
plt.legend(fontsize = 12, loc = 'upper left')
plt.title('Scatter Plot(iris)', fontsize = 15)
plt.xlabel('Petal Length', fontsize = 12)
plt.ylabel('Petal Width', fontsize = 12)
plt.show()
```

```python
e = sns.catplot(x = 'sepal_width', y = 'sepal_length', col = 'species', data=iris, kind='swarm')
e.set(xticklabels=[])

e = sns.catplot(x = 'petal_width', y = 'petal_length', col = 'species', data=iris, kind='swarm')
e.set(xticklabels=[])
```

```python
sns.pairplot(data=iris, hue='species')
plt.suptitle('iris pair plot')
plt.show()
```

```python
co = iris.species.astype('category')
co = co.cat.codes
pd.plotting.scatter_matrix(iris, s= 100, c=co, cmap='viridis', figsize = (12,12))
plt.show()
```
