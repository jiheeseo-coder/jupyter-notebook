pyplot의 다양한 plot 함수

```python
#데이터 프레임의 값을 이용하여 다양한 그래프를 subplotting
df = pd.DataFrame({'label':['A', 'B', 'C', 'D'], 'values':[10,30,60,20]})
plt.figure(1, figsize=(9,3))
plt.subplot(131)
plt.bar(df['label'], df['values'])

plt.subplot(132)
plt.scatter(df['label'], df['values'])

plt.subplot(133)
plt.pie(df['values'], labels=df['label'], autopct='%1.1f%%')
plt.show()
```

```python
# 단일값 그래프
df1 = pd.DataFrame({'label' : ['A', 'B', 'C'], 'value':[10,30,20]})
display(df1)
#plot.bar()를 이용하여 막대그래프 그리기
#x : 축 레이블, y : 값 (범례), title:제목, rot: x축 출력값 기울기
df1.plot.bar(x='label', y='value', title='Bar plot', rot=0)
plt.show()
```

Multi-set Bar Chart (다중집합 막대 그래프)

    Grouped Bar Chart, Clustered Bar Chart
    막대 차트 변형으로 둘 이상의 데이터 셋을 표시하고 범주 아래에 모두 같은 축에 그룹화 될 때 사용

```python
#다중집합 막대그래프
speed=[0.1, 17.5, 40, 48,52,69,88]
lifespan = [2,8,70,1.5,25,12,28]
index=['snail', 'pig', 'elephant', 'rabbit', 'giraffe', 'coyote', 'horse']
df2 = pd.DataFrame({'speed' : speed, 'lifespan' : lifespan}, index = index)
display(df2)
df2.plot.bar(rot=0)
plt.show()

#세부속성 y값 순서, x축 값 기울기 조정
df2.plot.bar(y=['lifespan', 'speed'], rot=45)
plt.show()
```

```python
#단일 y값 출력
df2.plot.bar(y=['speed'], rot=0)
plt.show()

#subplot으로 분리된 그래프, # 서브플롯 true, 그림크기 (폭 8, 높이 4)
axes = df2.plot.bar(rot=0, subplots=True, figsize=(8,4))
plt.show()
```

```python
#수평막대 그래프, x축 lifespan, y축 index
df2.plot.barh(y='lifespan', title='lifespan barh plot', rot=0)
plt.show()

#stacked bar
df2.plot.bar(stacked=True, title="stacked bar plot")
plt.show()
```

Heatmap plot
    
    데이터 값을 색상에 매핑하는 시각화 방법
    테이블 형식의 행과 열에 변수를 배치하고 테이블 내의 셀에 데이터 값에 해당하는 색상을 지정
    히트맵과 색상범례(색상에 의존하여 값을 전달하기 때문) 함께 표시
    plt.pcolor()함수를 이용하여 히트맵을 그리고, plt.colorbar()로 색상에 대한 정보를 오른쪽에 표시

```python
#(1) "weaterTemp.csv"을 데이터프레임 temp로 로딩, 0번째 칼럼을 인덱스로 사용
temp = pd.read_csv('./data/weatherTemp.csv', index_col=0)

#(2) 기온 데이터의 정보확인 (개수, index, columns, column type등)
print("temp.info = ")
print(temp.info())
print()

#(3) 기온 데이터의 확인
print("temp = ", temp)
```

```python
#(4) 월별 각 지역 기온 plotting
temp.plot(kind='line', title='지역별 평년기온(1981-2010)')
plt.show()
```

```python
# 월을 x축으로 지역을 y축으로 하는 히트맵 그리기
#pyplot의 hitmap을 그리기 위해 plt.pcolor()이용
#plt.pcolor() : 데이터프레임의 columns이 X축 index가 y축으로 설정
#temp를 행렬변환(transpose)로 temp2 생성
temp2 = temp.transpose()
display(temp2)
#1월 기온 역순으로 정렬
temp2 = temp2.sort_values(by=1, ascending = False)
display(temp2)
#plt.pcolor()을 이용하여 temp2 heatmap plotting, colormap=plasma
plt.pcolor(temp2, cmap='plasma')   #X축 : columns(월), Y축 : index(지역)
plt.yticks(np.arange(0.5, len(temp2.index), 1), temp2.index)
plt.xticks(np.arange(0.5, len(temp2.columns), 1), temp2.columns)
plt.title('지역별 평년 기온', fontsize = 20)
plt.xlabel('월', fontsize=14)
plt.ylabel('지역', fontsize=14)
plt.colorbar()
plt.show()
```

Seaborn

    Matplotlib 위에 구축된 그래픽 라이브러리
    다양한 플롯 스타일 및 색상 테마 제공
    통계용 플롯 유형을 추가한 고급 기능 제공

```python
import seaborn as sns
#heatmap by seaborn
ax = sns.heatmap(temp2)
plt.title('Heatmap by seaborn', fontsize=20)
plt.show()
```

```python
#각 셀에 숫자 값으로 주석을 추가
sns.heatmap(temp2, annot=True)
plt.show()
```

Seaborn 에서 제공되는 데이터셋을 사용한 예

Anscombe's quartet

    기술통계량은 유사하나 분포가 다른 4개의 데이터 셋
    import() 함수를 이용하여 각 데이터 셋에 대한 scatter plot과 선형회귀로 얻은 직선을 확인
   
   

```python
import seaborn as sns_a
sns_a.set_theme(style="ticks")

#load the example dataset for Anscombe's quartet
df = sns_a.load_dataset("anscombe")

#Show the results of a linear regression within each dataset
sns_a.lmplot(x="x", y="y", col='dataset', hue='dataset', data=df,
             col_wrap=2, ci=None, palette='muted', height=4,
             scatter_kws={'s':50, 'alpha':1})
```
