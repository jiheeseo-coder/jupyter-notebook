```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.DataFrame(3* np.random.rand(4,2), index = ['a', 'b', 'c', 'd'], columns = ['x', 'y'])
display(df)
```

```python
#x 그래프
df.plot.pie(y='x', title='pie plot (x)')
#y 그래프
df.plot.pie(y='y', title='pie plot (y)', autopct = '%.2f%%')
```

```python
df.plot.pie(title='pie subplot', subplots = True, figsize = (8,4))
```
