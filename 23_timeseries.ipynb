```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
import pandas_datareader as pdr
from datetime import datetime

start = datetime(2019, 11, 1)
end = datetime(2020,11,20)

#삼성전자 종목코드(005930.KS)
df = pdr.DataReader('005930.KS', 'yahoo', start, end)
display(df.info())
display(df.head())
```

```python
# Individual time series
df.plot(kind = 'line', y = 'Close', title='samsung stock(2020)', rot=45)
plt.show()
```

```python
df.plot(kind = 'line', y = 'Volume', title='samsung stock(2020)', rot=45)
plt.show()
```

```python
#Multiple time series (pandas dataframe)
ax = df.plot(kind = 'line', y = 'Close', title='samsung stock(2020)', rot=45, figsize=(10, 8))
df.plot(kind = 'line', y = 'Open', ax = ax)
plt.show()
```

```python
#최근 데이터 30개만 추출
df2 = df.iloc[-30:]
display(df2.info())
display(df2.head())
display(df2.tail())
```

```python
## Multiple time sries (pyplot)
plt.plot(df2.index, df2.Close, label = 'Close')
plt.plot(df2.index, df2.High, label = 'High')
plt.plot(df2.index,df2.Low, label = 'Low')
plt.legend()
plt.xticks(rotation = 45)
plt.show()
```
